```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url = 'https://bina.az/?gclid=Cj0KCQjw9MCnBhCYARIsAB1WQVWtSNmofbgLq38gmfDKpnrsu_cunaYcDfV37c_wLXqHRpQ80zMxBIcaAsssEALw_wcB'
source = requests.get(url)
source
```

```python
soup = BeautifulSoup(source.text , 'html.parser')
soup
```

```python
price = []
text = soup.find_all('div' , {'class':'price'})
for i in text:
    price.append(i.text)

price
```

```python
date_time = []
city = []
pattern = r'<div class="city_when">[^,]+, ([^<]+)</div>'
pattern2 = r'<div class="city_when">([^,]+), [^<]+</div>' # select city
text = soup.find_all('div' , {'class':'card_footer'})
text
for i in text:
    split_date = re.findall(pattern , str(i))
    split_city = re.findall(pattern2 , str(i))
    date_time.append(split_date)
    city.append(split_city)
print(date_time)
print(city)
```

```python
links = []
for i in soup.find_all('a' , {'class':'item_link'}):
    links.append('https://bina.az/' + i.get('href'))
links
```

```python
locations = []
for i in soup.find_all('div' , {'class':'location'}):
    locations.append(i.text)
locations
```

```python
location_propetry = zip(city , locations)
```

```python
dict_data = {'Location':location_propetry,
             'Links':links[0:40],
             'Date and Time': date_time,
             'Price': price}
document = pd.DataFrame(dict_data)
document
#I coudnt show in DataFrame format, because not all lists ar the same length
```
